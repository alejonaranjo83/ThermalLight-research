## 2.2. Todo integrado usando seaborn para graficar

```python
# TODO INTEGRADO Y USANDO SEABORN:

# Genero una tabla cuyo index sean los 7 votos térmicos posibles,
# las columnas sean los rangos de Lux y los valores provengan de
# Preferencia y Sensación

votos_pivot = votos_termicos_original.pivot_table(index="Voto térmico (no normalizado)",
                                                  columns="RangoLux",
                                                  values = ("Sensación_Cant_votos",
                                                            "Preferencia_Cant_votos"))
votos_pivot

#Configuro el tipo de apariencia q quiero para el gráfico
sns.set(style="whitegrid")

#Configuro formato del gráfico
f, ax = plt.subplots(figsize=(5,5))

# Aplico "melt" para poder hacer gráfico con PREFER y SENSAC:
# Al hacer "melt" no me reconoce la escala liker q tenía como index,
# por lo cual más adelante tendré que hacer un proceso adicional
# para tenerla en DF con el que haré el gráfico final
vot_piv_m = votos_pivot.melt()

# 2) renombraré columnas de DF para que posteriormente 
# al hacer gráfico me las reconozca seaborn
import numpy as np
vot_piv_m = vot_piv_m.rename(columns={np.nan: "opciones", "value": "cantidad"})
vot_piv_m.columns = vot_piv_m.columns.fillna("opciones")

# Genero una columna que solo tenga escala liker:
liker = votos_termicos_original
liker_m = liker.melt().head(35)

# Renombro columnas para poder trabajar después con ellas:
liker_m = liker_m.rename(columns={"value": "VotosTérmicos", "variable": "guia"})

# Necesito una columna donde los valores de liker se repitan 1 vez
# debajo de las filas previamente creadas:
double_liker = pd.concat([liker_m,liker_m])

# Cambio el index
double_liker.set_index("guia")


# Fusionando ambos DF´s:
# Aíslo solo la columna de liker
doubl_liker_alone = double_liker ["VotosTérmicos"]
doubl_liker_alone = double_liker.reset_index()
# Elimino columnas
p = doubl_liker_alone.drop(["guia", "index"], axis=1)

# Genero un nuevo DF con el q obtuve al hacer "melt" y el "liker" por aparte
final = pd.concat([p, vot_piv_m], axis=1)


# Hago gráfico:
diam = final["cantidad"]

ax.set_facecolor('none')
ax = sns.scatterplot(x="VotosTérmicos",
                     y="RangoLux", hue="opciones", 
                     #size="cantidad",
                     size=diam*20,
                     sizes=(1,2900),
                     facecolors='none',
                     edgecolor="k",
                     #legend=False,
                     alpha=0.4,
                     palette=["aquamarine", "dimgrey"],
                     data=final)
plt.ylim(0.3,5.5)
plt.xlim(-3.5,3.5)
plt.title("Preferencia y Sensación térmica según rangos de Lux", fontsize=12)

plt.legend(bbox_to_anchor=(1.1, 1), loc='upper left', borderaxespad=0)


# En estos links se explican en detalle cosas q hay detrá de lo hecho previa/:
# https://seaborn.pydata.org/generated/seaborn.scatterplot.html?highlight=scatterplot#seaborn.scatterplot
# https://matplotlib.org/3.2.1/tutorials/intermediate/legend_guide.html
```
