# *Análisis*

* *Al analizar el comportamiento de la sensación y preferencia térmica en cada una de las ciudades, se evidencia claramente el efecto de la adaptación del ser humano a las condiciones climáticas del lugar en que habita. Es así como un mismo valor de temperatura, se interpreta diferente en términos de sensación y preferencia en cada una de las ciudades.*

* *Como ilustración de lo anteriormente expuesto, se puede ver como cuando la temperatura ronda los 25ºC, la sensación térmica en Armenia es "muy fría" (-3), en Medellín está cercana a percibirse como "agradable" (0) tendiendo a "ligeramente fría" (-1), y en Manizales a sentirse "agradable" con tendencia hacia  "ligeramente caliente" (+1); por el contrario, con ese mismo valor de temperatura de 25ºC , la preferencia en Manizales se desplaza hacia temperaturas más bajas (-1, -2 y -3), en Medellín se puede mantener en el mismo valor (0) y en Armenia prefieren mayores temperaturas (+2).*

* *Pese a las condiciones son diferentes entre las diferentes ciudades, se observan tendencias similares en la variación de los votos de sensación preferencia en las tres ciudades.*

* *En Manizales nunca hubo votos de sensación correspondientes a "muy frío" (-3), ni votos de preferencia por un ambientes "muy caliente" (+3). En Armenia por su parte, no hubo votos de preferencia iguales a "muy caliente" (3). Medellín fue la única ciudad en la que se encontraron votos en cada una de las opciones ofrecidas en la escala liker adoptada en estudios de confort térmico; esto puede deberse a una mayor variabilidad en las condiciones climáticas de la ciudad, con un mayor contraste de períodos fríos y cálidos.*

* *Cuando la temperatura desciende, la sensación de las personas de Medellín tiende a ser más cercana a la que se observa en Manizales, mientras cuando se eleva, la sensación se asemeja más a la de las personas de Armenia. En ambos casos, la sensación y la preferencia térmica de las personas de Medellín, sugiere una mayor tolerancia por parte de las mismas hacia momentos de frío y calor. Si bien en Armenia se demoran más en expresar sensaciones de calor comparado con Medellín, cuando la temperatura desciende, son mucho más sensibles a experimentar frío que los de Medellín. Lo anterior puede verse también en Manizales: cuando la temperatura desciende, se demoran más en expresar sensaciones de frío comparado con los de Medellín, sin embargo cuando la temperatura aumenta, son mucho más sensibles y experimentan calor más rápido que los de Medellín.*

* *Un comportamiento similar al descrito anteriormente para los votos de "sensación", puede verse en los de "preferencia", lo cual podría dar indicios de una zona de confort más amplia en las personas de Medellín, comparada con los rangos de Armenia y Manizales. Este comportamiento indica que en condiciones donde las personas se sienten menos a gusto con su ambiente lumínico, estas preferían realizar modificaciones mayores sobre el comportamiento lumínico del espacio, comparado con las que se observan en las filas superiores. Dicho de otra manera, cuando el ambiente lumínico es más desfavorable, las personas parecen ser más sensibles al aspecto visual que el térmico.*
